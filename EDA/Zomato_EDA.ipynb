```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

### We have countrycode in the two dataframes hence the need to combine the two dataframes

```python
# combining the two df
## will perform an left join because our focus is on the 1st dataset
final_df=pd.merge(df,df_country,on = 'Country Code', how ='left')
```

```python
final_df.head()
```

```python
## Getting the datatype infromation
final_df.dtypes
```

```python
country_names=final_df['Country'].value_counts().index
```

```python
country_values = final_df['Country'].value_counts().values
```

```python
## Pie Chart (Top 3 countries that use zomato)
plt.pie(country_values[:3],labels= country_names[:3],autopct='%1.2f%%')
```

### Observation
    Zomato is dominant in India followed by United States and then the United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

### Observations
    1.When rating is between 4.5 to 4.9 ---> Excellent
    2.When rating is between 4.0 to 4.4 ---> Very Good
    3.When rating is between 3.5 to 3.9 ---> Good
    4.When rating is between 2.5 to 3.4 ---> Average
    5.When rating is between 2.0 to 2.4 ---> Poor

```python
#visualize the ratings
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating',y='Rating Count',data = ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue ='Rating color',data = ratings, palette = ['Blue','red','orange','yellow','green','green'])
```

### Observation
    1.Not rated count is very high.
    2.Maximum number of ratings are between 2.5 to 3.4

```python
#Frequency of the Rating color in the data frame
sns.countplot(x='Rating color',data=ratings,palette= ['Blue','red','orange','yellow','green','green'])
```

```python
#countiries that gave 0 rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

### Observation
    Maximum number of zero ratings are from indian customers.

```python
#Finding the currency used in the countries.
final_df[['Currency','Country']].groupby(['Currency','Country']).size().reset_index()
```

```python
## Which countries have online deliveries?
final_df[['Country','Has Online delivery']].groupby(['Country','Has Online delivery']).size().reset_index()
```

### Observation
    Online deliveries are available in India and UAE
